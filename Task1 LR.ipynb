### Preparing the Data

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 6 )
```

### Training the Model

```python
model = LinearRegression()
model = model.fit(X_train,y_train)
```

```python
# fitting the regression line to the points

line = model.coef_*X+model.intercept_

plt.scatter(X,y)
plt.plot(X,line, c='r', label='regression line')
plt.xlabel('Hours studied')
plt.ylabel('Percentage Scored')
plt.title('Hours vs Percentage')
plt.legend()
plt.show()
```

### Predicting and Evaluating the Model

```python
y_pred = model.predict(X_test)
```

```python
actualvpred = pd.DataFrame({'Actual':y_test, 'Predicted': y_pred})
```

```python
actualvpred
```

```python
plt.plot(y_pred, c='pink', label= 'Predicted')
plt.plot(y_test, c='b', label = 'Actual')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Scored')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
```

```python
print('Mean Absolute Error: ', mae, '\nMean Squared Error: ', mse)
```

Evaluating the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. For simplicity here, we have chosen the mean square error and mean absolute error.
